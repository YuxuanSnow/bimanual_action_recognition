```python
import imageio
import os
import matplotlib.pyplot as plt
import cv2
import torch
import torchvision
import json
```

```python
# load one frame image
base_path = "/home/yuxuan/project/Bimanual_Action_Recognition/KIT_BIMACS_DATASET/"
subject_task_take = os.path.join("subject_1", "task_1_k_cooking",  "take_1")
frame = "frame_100.png"

image_path = os.path.join(base_path, "bimacs_rgbd_data", subject_task_take, "rgb", "chunk_1", frame)
assert os.path.exists(image_path)
original_img = cv2.imread(image_path)

standard_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB) # convert to normal colour
plt.imshow(standard_img)
```

```python
# visualize skeleton

with open(os.path.join(base_path, "bimacs_derived_data",subject_task_take,"body_pose","frame_100.json")) as f:
    body_pose = json.load(f)

x_body = []
y_body = []
for body_part in body_pose[0]:

    if body_pose[0][body_part]['confidence'] != 0:
        x_body.append(body_pose[0][body_part]['x'])
        y_body.append(body_pose[0][body_part]['y'])


for x1,y1 in zip(x_body,y_body):
    skeleton_image = cv2.circle(standard_img, (int(x1),int(y1)), radius=5, color=(255, 0, 0), thickness=-1)
    
plt.imshow(skeleton_image)
```

```python
# visualize hand
with open(os.path.join(base_path, "bimacs_derived_data",subject_task_take,"hand_pose","frame_100.json")) as f:
    hand_pose = json.load(f)

x_hand = []
y_hand = []
for hand_part in hand_pose[0]:

    if hand_pose[0][hand_part]['confidence'] != 0:  # only add when confident with prediction
        x_hand.append(hand_pose[0][hand_part]['x']*640)
        y_hand.append(hand_pose[0][hand_part]['y']*480)

for x1,y1 in zip(x_hand,y_hand):
    skeleton_hand_image = cv2.circle(standard_img, (int(x1),int(y1)), radius=2, color=(0, 255, 0), thickness=-1)

plt.imshow(skeleton_hand_image)

plt.savefig('skeleton.png')
```

```python
# read 2d object json file

with open(os.path.join(base_path, "bimacs_derived_data",subject_task_take,"2d_objects","frame_100.json")) as f:
    object_2d = json.load(f)

cropped_img = []
obj_label = []

for i in object_2d:
    
    x = i["bounding_box"]['x']*640
    y = i["bounding_box"]['y']*480
    h = i["bounding_box"]['h']*480
    w = i["bounding_box"]['w']*640

    y1 = int(y-h/2)
    y2 = int(y+h/2)
    x1 = int(x-w/2)
    x2 = int(x+w/2)
    
    croped = standard_img[y1:y2,x1:x2]
    
    cropped_img.append(croped)
    obj_label.append(i['candidates'][0]['class_name'])
    
plt.imshow(cropped_img[0])
obj_label[0]
```

```python
# extract features from roi

# operation on image
input_h, input_w = 224, 224

transform = torchvision.transforms.Compose([
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                         std=[0.229, 0.224, 0.225]),])

# network

action_class_number = 14
descriptor_length = 25

class Resnet152(torch.nn.Module):
    def __init__(self, num_classes=descriptor_length):
        super(Resnet152, self).__init__()
       
        # use resnet-152 (Deep Residual Learning for Inage Recognition)
        self.learn_modules = torchvision.models.resnet152(pretrained=True)

        # fully connected layer to do the classification
        self.visualize_feature = 0
        self.fc_ = torch.nn.Linear(1000, 200)
        self.fc = torch.nn.Linear(200, num_classes)
        # self.fc = torch.nn.Sequential(
        #     torch.nn.ReLU(True),
        #     # torch.nn.Dropout(),
        #     torch.nn.Linear(1000, num_classes),
        # )
        # self._initialize_weights()

    def forward(self, x):
        x = self.learn_modules(x)
        x = x.view(x.size(0), -1)
        x = self.fc_(x)
        output = self.fc(x)
        x=output
        return x, output

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, torch.nn.Linear):
                m.weight.data.normal_(0, 0.01)
                m.bias.data.zero_()

feature_network = Resnet152()
feature_network = torch.nn.DataParallel(feature_network).cuda()

feature_roi = []

for i in cropped_img:
    roi_image = transform(cv2.resize(i, (input_h, input_w), interpolation=cv2.INTER_LINEAR))
    roi_image = torch.autograd.Variable(roi_image.unsqueeze(0)).cuda()
    # unsqueeze because pytorch expect a batch of image as input

    feature, _ = feature_network(roi_image)

    feature_roi.append(feature.squeeze())
```

```python
# action_classifier.dataset.py


import os
import json
import pickle
import time
import traceback
from typing import List, Tuple, Dict, Generator, Any

import action_classifier.actclass as ac

def get_dataset_paths(dataset_basepath: str, dataset_config: str) -> List[str]:
    paths: List[str] = []

    for subject, task, take in crawl_dataset():
        rec_basepath = os.path.join(dataset_basepath, dataset_config, subject, task, take)
        frame_count: int = len(os.listdir(rec_basepath)) // 2
        for i in range(frame_count):
            paths.append(os.path.join(rec_basepath, 'frame_{}_left.cache'.format(i)))
            paths.append(os.path.join(rec_basepath, 'frame_{}_right.cache'.format(i)))

    paths.sort()

    return paths


def get_raw_dataset_paths(derived_data_basepath, ground_truth_basepath: str) -> Tuple[List[str], List[str]]:
    derived_data_paths: List[str] = []
    ground_truth_paths: List[str] = []
    # list contains path to the derived and groundtruth data

    for subject, task, take in crawl_dataset():
        derived_data_paths.append(os.path.join(derived_data_basepath, subject, task, take))
        ground_truth_paths.append(os.path.join(ground_truth_basepath, subject, task, take + '.json'))

    return derived_data_paths, ground_truth_paths


def crawl_dataset() -> Generator[Tuple[str, str, str], None, None]:
    for subject in ['subject_{}'.format(i) for i in range(1, 7)]:
        for task in ['task_1_k_cooking', 'task_2_k_cooking_with_bowls', 'task_3_k_pouring',
                     'task_4_k_wiping', 'task_5_k_cereals', 'task_6_w_hard_drive',
                     'task_7_w_free_hard_drive', 'task_8_w_hammering', 'task_9_w_sawing']:
            for take in ['take_{}'.format(take_i) for take_i in range(10)]:
                yield subject, task, take

d_path = "/home/yuxuan/project/Bimanual_Action_Recognition/KIT_BIMACS_DATASET/bimacs_derived_data"
g_path = "/home/yuxuan/project/Bimanual_Action_Recognition/KIT_BIMACS_DATASET/bimacs_rgbd_data_ground_truth"

dd_path, gd_path = get_raw_dataset_paths(d_path, g_path)

dataset_path = "/home/yuxuan/project/Bimanual_Action_Recognition/KIT_BIMACS_DATASET"
base_path = "/home/yuxuan/project/Bimanual_Action_Recognition"
```
